```python
import pandas as pd
import IPython
```

```python
#pd.set_option('display.notebook_repr_html', True)
# Series
s = pd.Series([2,4,8,6,4],index=['a','b','c','d','e'])
print(s)
```

```python
# dataframe
data = {'Country' : ['india','france','england'],
       'capital' : ['New delhi','Paris','London'],
       'Population' : [1303131035,111908046,204565786]}
df = pd.DataFrame(data,columns=['Country','capital','Population'])

print(df)
display(df)
```

```python
# I/O

# read and write csv
df.to_csv('country.csv',index=None)
df1 = pd.read_csv('country.csv')
display(df1)

# read and write from excel
df.to_excel('country.xlsx',sheet_name='country')
df2 = pd.read_excel('country.xlsx',sheetname='country')
display(df2)
```

```python
# Selection

# getting
display(s)
print(s['b'])

display(df)
display(df[:1])                    # :1 print everything before 1--- 1: print everything after 1

# by position
print(df.iloc[0,0],'\n')                     
print(df.iat[2,0],'\n')

# by label 
print(df.loc[0,'Country'],'\n')
print(df.at[1,'Country'],'\n')

print(df.ix[2],'\n')
print(df.ix[:,'capital'],'\n')
print(df.ix[1,'capital'],'\n')

# boolean indexing
print(s[~(s>3)],'\n')
print(s[(s<-1) | (s>2)],'\n')

# setting
s['a'] = 6
print(s)
```

```python
# Dropping
display(s)                                          #not working
s.drop('a',axis=0)
display(s)

display(df)
df.drop('Country',axis=1)
```

```python
# sort and rank
display(df.sort_index())
display(df.sort_values(by='Country'))
print(df.rank)
```

```python
# Data frame information

# basic information
print(df.shape,' = shape\n')
print(df.index,' = index\n')
print(df.columns,' = coloumns\n')
print(df.info(),' = info\n')
print(df.count(),' = count\n')

# Summary
print(df.sum(),' = sum of values\n')
print(df.cumsum(),' = cummulative sum of values\n')
print(df.min(),' = min\n')
print(df.max(),' = max\n')
#print(df.idxmin(),' = min index value\n')                                    # not working
#print(df.idxmax(),' = max index value\n')
print(df.describe(),' = describe\n')
print(df.mean(),' = mean\n')
print(df.median(),' = median\n')
```

```python
# applying functions

f = lambda x: x*2
display(df.apply(f))
display(df.applymap(f))
```

```python
# data allignment
s3 = pd.Series([9,2,5],index=['a','c','d'])
display(s3)
s3 = s + s3
display(s3)
display(s)

display(s.add(s3,fill_value=0))
print(s.sub(s3,fill_value=2),' = sub\n')
print(s.div(s3,fill_value=4),' = div\n')
```
