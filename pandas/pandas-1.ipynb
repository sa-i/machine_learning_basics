```python
# Reshape Data

data = {'date' : ['2017-04-01','2017-06-01','2017-03-01','2017-05-01','2017-03-01','2017-04-08'],
       'type' : ['a','b','c','a','a','b'],
       'value' : [11.32,65.32,45.26,85,65.12,13]}
df2 = pd.DataFrame(data)
display(df)

# pivote
df3 = df2.pivot(columns='type',index='date',values='value')

display(df3)
```

```python
# pivot table
df4 = pd.pivot_table(df2,columns='type',values='value',index='date')
display(df4)
```

```python
# stacking

stacked = df3.stack()
display(stacked)
stacked.unstack()
```

```python
# melt
display(df2)
print('\n')
display(pd.melt(df2,id_vars=['date'],value_vars=['type','value'],value_name='observations'))
```
